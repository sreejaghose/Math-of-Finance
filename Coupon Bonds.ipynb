```python
# We can use other libraries.
import math
#pip install numpy
import numpy as np
from matplotlib import pyplot as plt
```

```python
total_days = 360

# compounding period
m = 4 
m_days = total_days/m

# Coupon Rate
C = 0.03
# yield
y = 0.04
```

```python
def w(day):
    return (day % m_days)/m_days

def payments_left(day):
    return math.ceil((total_days - day)/m_days)

def dirty_price_factor(x):
    N = payments_left(x)
    _w = w(x)
    a = (1/(1+y/m))
    PV_face = a**(N-_w)
    PV_coup = [a**(i-_w) for i in range(1, N+1)]
    return PV_face + sum(PV_coup)*C/m

def accrued_interest(x):
    return w(x)*C/m

def clean_price_factor(x):
    return dirty_price_factor(x) - accrued_interest(x)
```

```python
X = np.linspace(0, total_days, total_days)
Y_1 = [dirty_price_factor(x) for x in X]
Y_2 = [clean_price_factor(x) for x in X]

plt.plot(X, Z_1, Z_2)
plt.legend(labels=["Dirty","Clean"])
```
