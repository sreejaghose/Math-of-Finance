```python
"""
We can assign values to variables as below.
In general variable names should tell you what the variable is,
for instance `principle` instead of `x`
"""
principle = 10000
rate = 0.18
n = 5
```

```python
# We can also assign the ouput of functions or aritmetic
annuity_factor = (1-(1+rate)**(-n))/rate
C = principle/annuity_factor

# If we want to see a value, we can print it.
print(f"The annual payment is: {C:,.2f}")

# We can also compute how much we will spend
print(f"We will spend a total: {C*n:,.2f}")
```

```python
balance = principle
interest = 0
payment_to_balance = 0
tot_int = 0
tot_pay = 0

# Formating is avaialble to print things like headers.
print(f"Year\t{'IntPart':>10}\t{'P_Part':>10}\t{'BalLeft':>10}\t|\t{'P_PaidTot':>10}\t{'IntPaidTot':>10}")
print("{}\t{:>10.3f}\t{:>10.3f}\t{:>10.3f}\t|\t{:>10.3f}\t{:>10.3f}".format(0, 0, 0, balance, 0, 0))

# This is a for loop, the start in inclusive, the end is exclusive,
# so this loop runs for 1,...,n
for i in range(1, n+1):
    # compute the interest on the remaining balance
    interest = balance*rate
    
    # Figure out how much of our payment goes to the balance
    payment_to_balance = C - interest
    
    # Reduce the balance, the -= syntax is a shortcut for `balance = balance - payment_to_balance`
    balance -= payment_to_balance
    
    # Let's keep a running total of the interest we have paid
    tot_int += interest
    tot_pay += payment_to_balance
    
    # Print the results.
    print(f"{i}\t{interest:>10,.3f}\t{payment_to_balance:>10,.3f}\t{balance:>10,.3f}\t|\t{tot_pay:>10,.3f}\t{tot_int:>10,.3f}")
```
